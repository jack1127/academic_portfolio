```python
import pandas as pd
import numpy as np
import pickle
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("vehicles_v2.csv")
df
```

```python
df = df.drop(['Unnamed: 0','region','posting_date'], axis=1)
df = df.dropna()
df
```

```python
df = df.drop(df[df.price == 0].index)
df
```

```python
print(np.percentile(np.array(df.price), 2.5))
print(np.percentile(np.array(df.price), 97.5))
```

```python
df = df.drop(df[df.price < 1450].index)
df = df.drop(df[df.price > 42990].index)
df
```

```python
y = np.array(df['price'])
x_cont = df[['year', 'odometer']]
x_nonc = df[['model','manufacturer','condition','cylinders','fuel','title_status','transmission','drive','type','paint_color','state']]
```

```python
dummy = pd.get_dummies(x_nonc)
dummy
```

```python
dummy = np.array(dummy)
Sum = np.sum(dummy, axis=0)
dummy = np.delete(dummy, np.where(Sum < 10)[0], axis=1)
```

```python
print(dummy.shape)
```

```python
x = np.concatenate((x_cont, dummy), axis=1)
print(x.shape)
```

```python
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(np.mean(y_train))
print(np.std(y_train))
print(np.mean(y_test))
print(np.std(y_test))
```

```python
x_subtrain, x_valid, y_subtrain, y_valid = train_test_split(x_train, y_train, test_size=0.2, random_state=0)
```

```python
x_stack, x_valid, y_stack, y_valid = train_test_split(x_valid, y_valid, test_size=0.5, random_state=0)
```

```python
print("x_subtrain shape = ", x_subtrain.shape)
print("x_valid shape = ", x_valid.shape)
print("x_stack shape = ", x_stack.shape)
print("x_test shape = ", x_test.shape)
```

```python
dataset = {'x_subtrain': x_subtrain, 'y_subtrain': y_subtrain, 'x_valid': x_valid, 'y_valid': y_valid, 'x_stack': x_stack, 'y_stack': y_stack, 'x_test': x_test, 'y_test': y_test}
```

```python
with open('dataset.pickle', 'wb') as fp:
    pickle.dump(dataset, fp)
```

```python
import xgboost as xg 
xgb_r = xg.XGBRegressor(objective ='reg:squarederror', n_estimators = 500, seed = 0) 
  
# Fitting the model 
xgb_r.fit(x_train, y_train) 
```

```python
# Predict the model 
pred = xgb_r.predict(x_test)
  
# RMSE Computation 
rmse = np.sqrt(mean_squared_error(y_test, pred)) 
print("rmse = ", rmse)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(weights='distance',n_neighbors=10,n_jobs=-1)
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, pred)) 
print("rmse = ", rmse)
```
